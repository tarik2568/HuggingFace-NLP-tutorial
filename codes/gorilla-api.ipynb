```python
import openai

openai.api_key = "EMPTY" # Key is ignored and does not matter
openai.api_base = "http://34.132.127.197:8000/v1"
```

```python
def get_gorilla_response(prompt= "I would like to extract sentiments from Bengali texts by fine-tuning.", model="gorilla-7b-hf-v0"):
    completion = openai.ChatCompletion.create(
      model=model,
      messages=[{"role": "user", "content": prompt}]
    )
    return completion.choices[0].message.content
```

```python
prompt = "I would like to extract sentiments from Bengali texts by fine-tuning."
print(get_gorilla_response(prompt, model="gorilla-7b-hf-v0" ))
```
