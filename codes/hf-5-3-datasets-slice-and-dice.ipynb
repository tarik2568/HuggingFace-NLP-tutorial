```python
def lowercase_title(example):
    return {"title": example['title'].lower(),"context": example['context'].lower(),"question": example["question"].lower(), "answers":example["question"].lower()}
    
```

```python
squad_lowercase = squad.map(lowercase_title)
```

```python
squad_lowercase.shuffle(seed = 42)["title"][:5]
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
```

```python
def tokenizer_title(example):
    return tokenizer(example["title"])
```

```python
squad_tokenized_title = squad.map(tokenizer_title, batched = True, batch_size = 500)
squad_tokenized_title
```

```python
squad.shuffle(seed = 42)["title"][:5]
```

```python
squad_tokenized_title.shuffle(seed = 42)["input_ids"][:5]
```

```python
def tokenizer_context(example):
    return tokenizer(example["context"])
```

```python
squad_tokenized_title_context = squad_tokenized_title.map(tokenizer_context, batched = True, batch_size = 500)
squad_tokenized_title_context
```

```python
squad_tokenized_title_context.shuffle(seed = 42)["input_ids"][:5]
```
