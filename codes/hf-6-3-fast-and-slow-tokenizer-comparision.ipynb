```python
from datasets import load_dataset
raw_datasets = load_dataset("glue","mnli")
raw_datasets
```

```python
from transformers import AutoTokenizer
fast_tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
```

```python
def tokenize_with_fast(examples):
    return fast_tokenizer(
    examples['premise'], examples['hypothesis'],truncation = True)
```

```python
slow_tokenizer = AutoTokenizer.from_pretrained("bert-base-cased",use_fast = False)
```

```python
def tokenize_with_slow(examples):
    return slow_tokenizer(
    examples['premise'], examples['hypothesis'],truncation = True)
```

```python
%%time
tokenized_dataset_fast = raw_datasets.map(tokenize_with_fast, batched = True, num_proc =8)
```

```python
%%time
tokenized_dataset_slow = raw_datasets.map(tokenize_with_slow, batched = True, num_proc =8)
```

```python
print(fast_tokenizer("Hello, how are you?")["input_ids"])
print(fast_tokenizer("Hello, how are              you?")["input_ids"])
```

```python
text = ["Let's talk about tokenizers superpowers.",
        "Fast tokenizer provide the fastest response possible."]
encoding = fast_tokenizer.encode_plus(text, return_offsets_mapping=True,  return_tensors="pt")

#print(encoding.sentence_ids())
print("Data type:",type(encoding))
print("Keys:",encoding.keys())
print("Tokens:", encoding.tokens())
print("Word IDs:", encoding.word_ids())
print("token_type_ids:", encoding["token_type_ids"])
print("Offset mapping:", encoding["offset_mapping"])
```

```python
print("Whether the tokenizer is fast? ---->", fast_tokenizer.is_fast)
print("Whether the tokenizer is fast? ---->", slow_tokenizer.is_fast)
```
