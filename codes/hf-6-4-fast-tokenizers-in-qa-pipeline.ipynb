```python
from transformers import pipeline
```

```python
qa_pipeline = pipeline("question-answering")
```

```python
context = """Bangladesh is still lagging behind competitors in the production of recycled yarn 
although the market for such yarn is swelling, especially in tune with new environmental laws being 
formulated by the European Union (EU). Recycled yarn is manufactured from worn-out or scrap textiles, 
furniture and clothing or other materials like plastic bottles. Though much of the recycling processes 
are patented, the basic concept involves creating pulp or fibres and turning those into yarn. 
Sometimes virgin cotton is also added to the mix. Use of apparel made from recycled yarn has been 
growing globally as consumers behaviour is changing centring the need for protecting the environment 
and slowing down climate change."""
```

```python
#question= "How recycled yarn is manufactured?"
question = "What is the capital of Bangladesh?"
```

```python
qa_pipeline(question = question, context = context)
```

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
```

```python
model_checkpoint = "distilbert-base-cased-distilled-squad"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
model = AutoModelForQuestionAnswering.from_pretrained(model_checkpoint)
```

```python
inputs = tokenizer(question, context, return_tensors = "pt")
outputs = model(**inputs)
```

```python
start_logits = outputs.start_logits
end_logits = outputs.end_logits
```

```python
print(start_logits.shape, end_logits.shape)
```

```python
 long_context = """
Bangladesh is a country in South Asia, located in the delta of the Padma (Ganges) and Jamuna (Brahmaputra) rivers in the northeastern part of the Indian subcontinent. It is the eighth-most populous country in the world, with a population of over 169 million people. The official language of Bangladesh is Bengali, and the capital city is Dhaka.

Bangladesh is a relatively young country, having gained its independence from Pakistan in 1971. It is a developing country, with a GDP per capita of \$1,870 (PPP). However, Bangladesh has made significant progress in recent years, and is now considered to be a lower-middle-income country by the World Bank.

Bangladesh is a unique country in many ways. It is the world's largest riverine delta, and is home to a wide variety of flora and fauna. The Sundarbans, a mangrove forest on the southern coast, is the largest mangrove forest in the world and is home to the endangered royal Bengal tiger.

Bangladesh is also a diverse country, with a rich culture and history. The people of Bangladesh are predominantly Muslim, but there are also significant Hindu, Buddhist, and Christian minorities. The country is home to a number of UNESCO World Heritage Sites, including the ruins of the ancient city of Mahasthangarh and the Sundarbans mangrove forest.

Here are some interesting facts about Bangladesh:

Bangladesh is the world's eighth-most populous country.
Bangladesh is the most densely populated country in South Asia.
Bangladesh is the world's largest riverine delta.
Bangladesh is home to the Sundarbans, the largest mangrove forest in the world.
Bangladesh is a Muslim-majority country.
Bangladesh is a developing country, but has made significant progress in recent years.
Bangladesh is a diverse country, with a rich culture and history.
I hope this gives you a brief overview of Bangladesh. If you have any other questions, please let me know.
"""
qa_pipeline(question=question, context=long_context)
```

```python
start_probabilities = torch.nn.functional.softmax(start_logits, dim=-1)[0]
end_probabilities = torch.nn.functional.softmax(end_logits, dim=-1)[0]
print(start_probabilities.shape)
print(end_probabilities.shape)
```

```python
result = {
    "answer": answer,
    "start": start_char,
    "end": end_char,
    "score": scores[start_index, end_index],
}
print(result)
```

```python
inputs = tokenizer(
    question,
    long_context,
    stride=128,
    max_length=384,
    padding="longest",
    truncation="only_second",
    return_overflowing_tokens=True,
    return_offsets_mapping=True,
)
```

```python
print(type(inputs))
print(inputs.keys())
print(inputs.n_sequences)
#print(inputs.items)
```

```python
_ = inputs.pop("overflow_to_sample_mapping")
offsets = inputs.pop("offset_mapping")

inputs = inputs.convert_to_tensors("pt")
print(inputs["input_ids"].shape)
```

```python
outputs = model(**inputs)

start_logits = outputs.start_logits
end_logits = outputs.end_logits
print(start_logits.shape, end_logits.shape)
```

```python
sequence_ids = inputs.sequence_ids()
# Mask everything apart from the tokens of the context
mask = [i != 1 for i in sequence_ids]
# Unmask the [CLS] token
mask[0] = False
# Mask all the [PAD] tokens
mask = torch.logical_or(torch.tensor(mask)[None], (inputs["attention_mask"] == 0))

start_logits[mask] = -10000
end_logits[mask] = -10000
```

```python
start_probabilities = torch.nn.functional.softmax(start_logits, dim=-1)
end_probabilities = torch.nn.functional.softmax(end_logits, dim=-1)
```

```python
candidates = []
for start_probs, end_probs in zip(start_probabilities, end_probabilities):
    scores = start_probs[:, None] * end_probs[None, :]
    idx = torch.triu(scores).argmax().item()

    start_idx = idx // scores.shape[1]
    end_idx = idx % scores.shape[1]
    score = scores[start_idx, end_idx].item()
    candidates.append((start_idx, end_idx, score))

print(candidates)
```

```python
for candidate, offset in zip(candidates, offsets):
    start_token, end_token, score = candidate
    start_char, _ = offset[start_token]
    _, end_char = offset[end_token]
    answer = long_context[start_char:end_char]
    result = {"answer": answer, "start": start_char, "end": end_char, "score": score}
    print(result)
```
